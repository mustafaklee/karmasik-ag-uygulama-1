```python
import networkx
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
```

```python
nodes_df = pd.read_csv('mlb_teams.csv', encoding='utf-16')
edges_df = pd.read_csv('mlb_games.csv',encoding='utf-16')
```

```python
nodes_df = nodes_df['prefix_1'].tolist()
```

```python
edges = list(zip(edges_df['away_team'], edges_df['home_team']))

#tekrar eden kenarlari silelim.
raw_edges= list(set(edges))
```

```python
#ynl bir a oluturalm
G = networkx.DiGraph()
#dmleri ekleyelim
G.add_nodes_from(nodes_df)
G.add_edges_from(raw_edges)
#A daha byk ve gzel izdirelim
plt.figure(figsize=(12,12))
networkx.draw(G, with_labels=True,node_size=50, font_size=10,font_color='black',node_color='lightblue')
plt.show()
```

```python
#dmlerin derecelerini hesalayalim
degrees = dict(G.degree())
sorted_degrees = dict(sorted(degrees.items(), key=lambda item: item[1], reverse=True))
print("Dmlerin Dereceleri (En Yksekten En De):")
for node, degree in sorted_degrees.items():
    print(f"{node}: {degree}")
```

```python
#ortalama derece hesaplayalim.
average_degree = sum(degrees.values()) / len(degrees)
print(average_degree)
```
