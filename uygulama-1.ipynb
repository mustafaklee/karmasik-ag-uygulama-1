```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
InpufileNodes = pd.read_csv("InputFileNodes.csv")
InputFileEdges = pd.read_csv("InputFileEdges.csv")
```

```python
#1-InputFileNodes dosyasnda ID stunu dm isimleri olacak.
nodes = InpufileNodes['id'].tolist()
#InputFileEdges dosyasnda from ve to stunlar kenarlar oluturacak
edges = list(zip(InputFileEdges['from'], InputFileEdges['to']))
#2- Bo bir grafik oluturma
G = nx.Graph()
#3- Dmleri ekleme
G.add_nodes_from(nodes)
#4- Kenarlar ekleme
G.add_edges_from(edges)
#5- Grafii izdirme
nx.draw(G, with_labels=True)
plt.show()
```

```python
#Az nce ynsz network oluturduk. imdi ynl network oluturalm.
#1- Bo bir ynl grafik oluturma
DG = nx.DiGraph()
#2- Dmleri ekleme
DG.add_nodes_from(nodes)
#3- Kenarlar ekleme
DG.add_edges_from(edges)
#4- Grafii izdirme
nx.draw(DG, with_labels=True)
plt.show()
```

```python
#G iin tm dmlerin dereceleri hesaplanacak
degrees = dict(G.degree())
print("Ynsz Graftaki Dmlerin Dereceleri:")
for node, degree in degrees.items():
    print(f"Dm {node}: Derece {degree}")
```

```python
#G network iin ortalama derece hesaplanacak
average_degree = sum(degrees.values()) / len(degrees)
print(f"\nYnsz Grafn Ortalama Derecesi: {average_degree:.2f}")
```

```python
#G networknn histogramn izdirelim.
degree_values = list(degrees.values())
plt.hist(degree_values, bins=range(min(degree_values), max(degree_values) + 2), align='left', rwidth=0.8)
plt.xlabel('Derece')
```

```python
#G network iin degrees centrality hesaplanacak
degree_centrality = nx.degree_centrality(G)
print("\nYnsz Grafn Derece Merkezilikleri:")
for node, centrality in degree_centrality.items():
    print(f"Dm {node}: Derece Merkezilik {centrality:.4f}")
```

```python
#G network iin Closeness centrality deerini hesaplayalim.
closeness_centrality = nx.closeness_centrality(G)
print("\nYnsz Grafn Closeness Merkezilikleri:")
for node, centrality in closeness_centrality.items():
    print(f"Dm {node}: Closeness Merkezilik {centrality:.4f}")
```

```python
#g network iin Eigenvector centrality deerini hesaplayalm.
eigenvector_centrality = nx.eigenvector_centrality(G)
print("\nYnsz Grafn Eigenvector Merkezilikleri:")
for node, centrality in eigenvector_centrality.items():
    print(f"Dm {node}: Eigenvector Merkezilik {centrality:.4f}")
```

```python
#Katz centrality hesaplayalim.
katz_centrality = nx.katz_centrality(G, alpha=0.1, beta=1.0)
```

```python
# PageRank centrality hesaplayalim.
pagerank_centrality = nx.pagerank(G, alpha=0.85)
print("\nYnsz Grafn PageRank Merkezilikleri:")
for node, centrality in pagerank_centrality.items():
    print(f"Dm {node}: PageRank Merkezilik {centrality:.4f}")
```

```python
#diamater hesaplayalm
diameter = nx.diameter(G)
print(f"\nYnsz Grafn ap: {diameter}")
```

```python
#derece dalmn izdirelim
degree_values = list(degrees.values())
plt.hist(degree_values, bins=range(min(degree_values), max(degree_values) + 2), align='left', rwidth=0.8)
plt.xlabel('Derece')
plt.ylabel('Dm Says')
plt.title('Ynsz Grafn Derece Dalm')
```

```python
#en kritik 5 kpr dm sralayalm
betweenness_centrality = nx.betweenness_centrality(G)
sorted_betweenness = sorted(betweenness_centrality.items(), key=lambda item: item[1], reverse=True)
print("\nYnsz Grafn En Kritik 5 Kpr Dm:",sorted_betweenness[:5])
```

```python
#en kritik 5 lider dm sralayalm
eigenvector_centrality = nx.eigenvector_centrality(G)
sorted_eigenvector = sorted(eigenvector_centrality.items(), key=lambda item: item[1], reverse=True)
print("\nYnsz Grafn En Kritik 5 Lider Dm:",sorted_eigenvector[:5])
```

```python
#DG icin derece dalmn hesaplayalm ve izdirelim
in_degrees = dict(DG.in_degree())
out_degrees = dict(DG.out_degree())
plt.hist(in_degrees.values(), bins=range(min(in_degrees.values()), max(in_degrees.values()) + 2), alpha=0.5, label='Gelen Derece', align='left', rwidth=0.4)
plt.hist(out_degrees.values(), bins=range(min(out_degrees.values()), max(out_degrees.values()) + 2), alpha=0.5, label='Giden Derece', align='left', rwidth=0.4)
plt.xlabel('Derece')
plt.ylabel('Dm Says')
plt.title('Ynl Grafn Derece Dalm')
```

```python
# g network icin en gl 3 dm bularak adan kaldralm ve yeni a izdirelim
sorted_degree = sorted(degrees.items(), key=lambda item: item[1], reverse=True)
top_3_nodes = [node for node, degree in sorted_degree[:3]]
G_removed = G.copy()
G_removed.remove_nodes_from(top_3_nodes)
nx.draw(G_removed, with_labels=True)
plt.show()
```
